First of all, we are considering this system with no external force and getting the equation of the Van der Pol equation is:
$$\ddot{x}-\mu\left(1-x^{2}\right) \dot{x}+x=0$$
To get started, we will write a code to solve the question analytically, by monitoring the value of $ \mu$,we will plot the solution of this unforced damping oscillation in a plot time series of x and $\dot{x}$, and phase space plots in the 2D space of (x, $\dot{x}$).
For convenience, we will set the initial value of distance and velocity as 0m and 1m/s.
The output of this part will be a 3-D graph and a section of phase space.

```python
import numpy as np #import libeary
from scipy.integrate import odeint#import libeary
import matplotlib.pyplot as plt#import libeary
from mpl_toolkits import mplot3d#import libeary

x0, v0= 0,1 #setting initial values
t = np.arange(0.0,200.001,0.001)#arrange the time array from 0 to 200 seconds, with a dt of 0.001 second

u=[1,2,3,4,4.5]#setting initial value of mu as initial value
for i in range(5):#start the for loop for the different value of mu
    
    def diff(y,t):
        '''define the ode function to let odeint to slove'''
        xo, vo = y#setting intial paramater for the function to slove
        dydt = [vo,u[i]*(1-xo**2)*vo-xo]#the ode function 
        return dydt#return the ode function to the odeint


    IC = [x0,v0]#the initial condtion 
    sol = odeint(diff,IC,t)#let the odeint to slove the function for us 
    
    # Set up a figure twice as tall as it is wide
    fig = plt.figure(figsize=[5,10])
    ax = fig.add_subplot(2, 1, 1, projection='3d')#first plt
    zline = sol[:, 0]#get the function of x
    xline = sol[:, 1]#get the function of v
    yline = t#the value of time varies
    ax.set_xlabel('distance(m)')#set label
    ax.set_ylabel('time(s)')#set label 
    ax.set_zlabel('velocity(m/s)')#set label
    ax.plot3D(xline, yline, zline, 'gray') #plot
    ax.set_title('mu =' + '%.3f' % u[i] + ' x0 =' + '%.3f'% x0 +' v0 =' + '%.3f'% v0+' 3D Phase space ')#title
    
    ax = fig.add_subplot(2, 1, 2)#second plot
    plt.title('mu =' + '%.3f' % u[i] + ' x0 =' + '%.3f'% x0 +' v0 =' + '%.3f'% v0+' x,v 2d Phase space')#title 
    plt.plot(sol[:, 0], sol[:, 1], 'gray')#plt
    plt.xlabel('x')#label
    plt.ylabel('v')#label
    plt.grid()#grid on
    
    plt.savefig('mu =' + '%.3f' % u[i] + ' x0 =' + '%.3f'% x0 +' v0 =' + '%.3f'% v0+'Phase spaces of undamped Van der Pol.png')
    #save fig

```

Now we will start to take the forced term and the equation will be transformed in the following form:
$$\ddot{x}-\mu\left(1-x^{2}\right) \dot{x}+x=A \cos (\omega t)$$
For this part, we will execute the same procedure as we operate in the above section, but during this time, we will try a few attempts of initial condition and make those values valid for the data of the next section.
The output of this part will be a 3-D graph and a section of phase space.
By the last section we will take $\mu$=3 in this case


```python
import numpy as np #import libeary
from scipy.integrate import odeint#import libeary
import matplotlib.pyplot as plt#import libeary
from mpl_toolkits import mplot3d#import libeary


t = np.arange(0.0,200.001,0.001)#arrange the time array from 0 to 200 seconds, with a dt of 0.001 second
x0 =[0,1,2,0,0]#setting ICs
v0 =[0,0,0,1,2]#setting ICs
phi0=0#setting ICs
mu= 3#setting ICs
A= 15#setting ICs
omega=4.20#setting ICs

for i in range(5):#start the for loop for the different value of mu
    
    def diff(y,t):
        '''define the ode function to let odeint to slove'''
        xo,vo,phio = y#setting ic
        dydt = [vo,mu*(1-xo**2)*vo-xo+A*np.cos(omega*t),omega]#ode equation
                
        return dydt #return the ode function to the odeint
                
    IC = [x0[i],v0[i],phi0]#the initial condtion 
    sol = odeint(diff,IC,t)#let the odeint to slove the function for us 
    
    # Set up a figure twice as tall as it is wide
    fig = plt.figure(figsize=[5,10])
    ax = fig.add_subplot(2, 1, 1, projection='3d')#first plt
    zline = sol[:, 0]#get the function of x
    xline = sol[:, 1]#get the function of v
    yline = t#the value of time varies
    ax.set_xlabel('distance(m)')#set label
    ax.set_ylabel('time(s)')#set label 
    ax.set_zlabel('velocity(m/s)')#set label
    ax.plot3D(xline, yline, zline, 'gray') #plot
    ax.set_title('mu =' + '%.3f' % mu + ' x0 =' + '%.3f'% x0[i] +' v0 =' + '%.3f'% v0[i]+' 3D Phase space ')#title
    
    ax = fig.add_subplot(2, 1, 2)#second plot
    plt.title('mu =' + '%.3f' % mu + ' x0 =' + '%.3f'% x0[i] +' v0 =' + '%.3f'% v0[i]+' x,v 2d Phase space')#title 
    plt.plot(sol[:, 0], sol[:, 1], 'gray')#plt
    plt.xlabel('x')#label
    plt.ylabel('v')#label
    plt.grid()#grid on
    
    plt.savefig('mu =' + '%.3f' % mu+ ' x0 =' + '%.3f'% x0[i] +' v0 =' + '%.3f'% v0[i]+'Phase spaces of damped Van der Pol.png')
    #save fig

    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
```

The 3D phase space (x, $\dot{x} ,\phi$ ) is difficult to visualize, even in projection. We will use a Poincare section which is a two-dimensional slice through higher-dimensional phase space to view the dynamics. Each time the trajectory intersects with the surface, it will make a point. In a periodic system, the motion repeats itself after a certain amount of time so the Poincare section is made up of a  number of discrete points.We will apply such section to the pervious section.\

```python
from scipy import interpolate#import libeary
from scipy import optimize#import libeary
from numpy import *#import libeary
from scipy.integrate import odeint#import libeary
import math#import libeary
from pylab import *#import libeary


def poincaresection(x,v,phi,t,omega,phimod2pi):
    '''use this function to computes the value of poincare section '''
    xinterp = interpolate.UnivariateSpline(t, x, s=0, k=3)#smoothen the points to a curve
    vinterp = interpolate.UnivariateSpline(t, v, s=0, k=3)#smoothen the points to a curve
    
    tmax=t[-1] #limiting the time
    
    numzero=int(floor((tmax*omega)/(2*math.pi)))#round angles to an integer 
        
    phiroot=0#setting strating value for phi
    i=1#setting strating value
    
    xpoincare=zeros(numzero)#empty the list
    vpoincare=zeros(numzero)#empty the list
    
    while True:
        phiroot=2*math.pi*i/omega+phimod2pi/omega#compute phi root
        
        if phiroot>tmax:#if more than the time stop the while loop
            break
            
        xpoincare[i-1]=xinterp(phiroot)#wirting poincare section data to the list
        vpoincare[i-1]=vinterp(phiroot)#wirting poincare section data to the list
        i=i+1#go to the next one 

    cutoff = int(numzero/2)#normalize for the poincare section
    return [xpoincare[cutoff:],vpoincare[cutoff:]]#return poincare section
    
#ic

x0 =2#setting IC for distance 
v0 =0#setting IC for velocity 
phi0=0#setting IC for phi
mu= 3#setting IC for mu
A= 15#setting IC for A
dt= 0.001 #setting IC for delta time
t = arange(0.5,1000,dt)#setting IC for time
phic =2#This is a number between 0 and 2pi which specifies the coordinate where the Poincare section should be plotted
omega=4.07#setting IC for omega


def diff(y,t):
    '''define the ode function to let odeint to slove'''
    xo,vo,phio = y#setting ic
    dydt = [vo,mu*(1-xo**2)*vo-xo+A*np.cos(omega*t),omega]#ode equation
                
    return dydt #return the ode function to the odeint
                
IC = [x0,v0,phi0]#the initial condtion 
sol = odeint(diff,IC,t)#let the odeint to slove the function for us 
    

xr, vr = poincaresection(sol[:, 0],sol[:, 1],phi0,t,omega,phic)#mark points

plt.plot(sol[:, 0], sol[:, 1])#plot the ode
plt.plot(xr, vr,'or')#plot the poincare
plt.grid()#grid on
suptitle('omega =' + '%.3f' % omega +' phi sec ='+ '%.3f' % phic)#title
#show()
savefig('omega ='+'%.3f' % omega + ' phi sec='+'%.3f' % phic + '.png')#save fig


```
